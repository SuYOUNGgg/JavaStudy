```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
age = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
# 0~25, 25~50, 50~75, 75~100
plt.hist(age, bins=[1,25,50,75,100])
plt.title("age result")
plt.xticks([0,25,50,70,100])
plt.show()
```
